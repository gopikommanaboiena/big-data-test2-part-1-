```python
import pandas as pd  
data = {'Number of neurons': [200,400,600,800], 'Time Difference': [23674,21860,12,12740]}  
df_H = pd.DataFrame(data)  
print(df_H)
```

```python
import matplotlib.pyplot as plt


# plot
plt.plot(df_H['Number of neurons'],df_H['Time Difference'])
plt.xlabel('Number of neurons')
plt.ylabel('Time taken')
plt.title('Learning rate is constant and Number of neurons is changing')
plt.grid(True)
  
plt.show()
```
